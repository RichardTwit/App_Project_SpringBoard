```python

import pandas as pd
import matplotlib.pyplot as plt

graph_columns=["close", "open", "high", "low",
 'MFI', 'MACD', 'Signal Line' ,
       'RSI_14', 'EMA_14', 'close_diff',
       'open_diff', 'high_diff', 'low_diff']

for column in graph_columns:
    plt.figure(figsize=(12, 6))
    plt.plot(df['date'], df[column], label='Time Series Data', color='blue')
    plt.title(column + ' Time Series Plot')
    plt.xlabel('Date')
    plt.ylabel(column)
    plt.legend()
    plt.grid(True)
    plt.show()
```

Close, open, high, low shows a downtrend.

```python

import pandas as pd
import matplotlib.pyplot as plt

columns = ["close", "open", "high", "low"]

for column in columns:
  df[column+ '_Lagged'] = df[column].shift(1)

  plt.figure(figsize=(8, 6))
  plt.scatter(df[column], df[column +'_Lagged'])
  plt.xlabel(column + ' (t)')
  plt.ylabel(column + ' (t-1)')
  plt.title(column + ' Lagged Scatter Plot')
  plt.show()
```

```python

import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

columns = ["close", "open", "high", "low"]

for column in columns:
  # Plot the ACF
  plt.figure(figsize=(8, 6))
  plot_acf(df[column], lags=40)
  plt.xlabel(column)
  plt.ylabel('ACF')
  plt.title('Autocorrelation Function (ACF) Plot')
  plt.show()
```

```python

import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_pacf

columns = ["close", "open", "high", "low"]

for column in columns:
  # Plot the PACF
  plt.figure(figsize=(8, 6))
  plot_pacf(df[column], lags=40)
  plt.xlabel(column)
  plt.ylabel('PACF')
  plt.title('Partial Autocorrelation Function (PACF) Plot')
  plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

columns = ["close_diff", "open_diff", "high_diff", "low_diff"]

for column in columns:
  df[column+ '_Lagged'] = df['close_diff'].shift(1)
# Plot the lagged scatter plot
  plt.figure(figsize=(8, 6))
  plt.scatter(df[column], df[column +'_Lagged'])
  plt.xlabel(column + ' (t)')
  plt.ylabel(column + ' (t-1)')
  plt.title(column + ' Lagged Scatter Plot')
  plt.show()
```

All four graphs are stationary becuase there is no clear trend and points are random. This is good for Arima model.
