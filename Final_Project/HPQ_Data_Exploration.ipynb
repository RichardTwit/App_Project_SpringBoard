```python
candle_columns= ['CDL3LINESTRIKE', 'CDLDOJI', 'CDLGRAVESTONEDOJI', 'CDLHARAMI',
       'CDLHARAMICROSS', 'CDLHIKKAKEMOD', 'CDLHOMINGPIGEON',
       'CDLINVERTEDHAMMER', 'CDLLONGLEGGEDDOJI', 'CDLMATCHINGLOW',
       'CDLRICKSHAWMAN', 'CDLSTICKSANDWICH']
```

```python
import pandas as pd
import matplotlib.pyplot as plt




#df['date'] = pd.to_datetime(df['date'])


#df.set_index('date', inplace=True)


columns_to_plot = ['open', 'close', 'high', 'low']


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))


for i, ax in enumerate(axes.flatten()):
    column = columns_to_plot[i]
    df[column].plot(ax=ax, linestyle='-', marker='o', markersize=4)
    ax.set_title(column.capitalize())


plt.tight_layout()


fig.suptitle('Trading Data Time Series')
fig.text(0.5, 0.04, 'Date', ha='center')
fig.text(0.04, 0.5, 'Price', va='center', rotation='vertical')


plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go


df.reset_index(inplace=True)

df['date'] = pd.to_datetime(df['date'])


fig = go.Figure(data=[go.Candlestick(x=df['date'],
                                     open=df['open'],
                                     high=df['high'],
                                     low=df['low'],
                                     close=df['close'])])


fig.update_layout(title='Trading Data Candlestick Chart',
                  xaxis_title='Date',
                  yaxis_title='Price')


fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt




value_counts = df[candle_columns].apply(pd.value_counts)

print(value_counts)


fig, ax = plt.subplots()

value_counts.plot(kind='bar', ax=ax)


plt.title('Value Counts of Columns')
plt.xlabel('Columns')
plt.ylabel('Count')


plt.xticks(rotation=90)


plt.show()
```
