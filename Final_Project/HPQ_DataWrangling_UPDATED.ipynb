```python


metric = 'volume'

# Find The Outlier in the data
mean = df[metric].mean()
std_dev = df[metric].std()


threshold = 3


outliers = df[(df[metric] > mean + threshold * std_dev) | (df[metric] < mean - threshold * std_dev)]


print("Outliers:")
print(outliers)
```

```python

metric = 'volume'


#Remove the outliers of the data
threshold = 3


mean = df[metric].mean()
std_dev = df[metric].std()


df_no_outliers = df[(df[metric] >= mean - threshold * std_dev) & (df[metric] <= mean + threshold * std_dev)]


print(df_no_outliers)
```

```python
column = 'volume'

# Calculate the mean of the column with outliers
mean_with_outliers = df[column].mean()


mean_without_outliers = df_no_outliers[column].mean()


print("Mean with outliers:", mean_with_outliers)
print("Mean without outliers:", mean_without_outliers)
```

```python
import talib
candle_names = talib.get_function_groups()['Pattern Recognition']
```

```python
candle_names
```

```python
df['Combined_Signal'] =0

for candle in candle_names:

    candle_function = getattr(talib, candle)


    signal = candle_function(df['open'], df['high'], df['low'], df['close'])


    df['Combined_Signal'] += signal


df['Trading_Signal'] = 'Hold'
df.loc[df['Combined_Signal'] > 0, 'Trading_Signal'] = 'Buy'
df.loc[df['Combined_Signal'] < 0, 'Trading_Signal'] = 'Sell'
```

```python

# Calculate Money Flow Index (MFI)
df['MFI'] = talib.MFI(df['high'], df['low'], df['close'], df['volume'], timeperiod=14)
```

```python

# Calculate MACD with fast length = 12 and slow length = 26
ema_fast = df['close'].ewm(span=12, adjust=False).mean()
ema_slow = df['close'].ewm(span=26, adjust=False).mean()
df['MACD'] = ema_fast - ema_slow
df['Signal Line'] = df['MACD'].ewm(span=9, adjust=False).mean()
```

```python


# Calculate the RSI with a window of 14 periods
delta = df['close'].diff()
gain = delta.mask(delta < 0, 0)
loss = -delta.mask(delta > 0, 0)
average_gain = gain.rolling(window=14).mean()
average_loss = loss.rolling(window=14).mean()
rs = average_gain / average_loss
df['RSI_14'] = 100 - (100 / (1 + rs))
```

```python

# Calculate an exponential moving average (EMA) with a span of 14 periods
df['EMA_14'] = df['close'].ewm(span=14, adjust=False).mean()
```

```python

from statsmodels.tsa.stattools import adfuller


alpha = 0.05


non_stationary_columns = []


for column in df.columns:
    column_values = df[column].values.flatten()
    result = adfuller(column_values)
    print(f"Column: {column}")
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])


    if result[1] <= alpha:
        print("Stationary (Reject the null hypothesis)\n")
    else:
        print("Non-Stationary (Fail to reject the null hypothesis)\n")


        non_stationary_columns.append(column)
```
